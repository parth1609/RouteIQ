```python
import csv
import json

def csv_to_json_no_index(csv_filepath, json_filepath):
    data = []
    with open(csv_filepath, 'r', encoding='utf-8') as csv_file:
        # csv.DictReader reads the first row as headers and maps subsequent rows to dictionaries
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            # If there's an empty string key (common for unnamed first column in some CSVs)
            # you can explicitly remove it here if it appears, although DictReader usually prevents it
            # if the header isn't actually blank in the CSV itself.
            if "" in row:
                del row[""] # This line removes the empty key if it exists
            data.append(row)

    with open(json_filepath, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, indent=4)

    print(f"CSV data from '{csv_filepath}' successfully converted to JSON in '{json_filepath}' without an index column.")

# Usage
csv_file = 'IT Support Ticket Data.csv'
json_file = 'IT Support Ticket Data.json'
csv_to_json_no_index(csv_file, json_file)
```
